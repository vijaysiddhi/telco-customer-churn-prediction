# Step 4: Feature Engineering

```python
# Create tenure groups
def tenure_group(tenure):
    if tenure <= 12:
        return '0-1 Year'
    elif tenure <= 24:
        return '1-2 Years'
    elif tenure <= 36:
        return '2-3 Years'
    elif tenure <= 48:
        return '3-4 Years'
    elif tenure <= 60:
        return '4-5 Years'
    else:
        return '5+ Years'

df['tenure_group'] = df['tenure'].apply(tenure_group)
```

```python
# Visualize churn by tenure groups
plt.figure(figsize=(10, 6))
sns.countplot(x='tenure_group', hue='Churn', data=df)
plt.title('Churn by Tenure Groups')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a feature for contract type (simplified)
df['Contract_Simple'] = df['Contract'].map({
    'Month-to-month': 'Short',
    'One year': 'Medium',
    'Two year': 'Long'
})
```

```python
# Visualize churn by simplified contract
plt.figure(figsize=(8, 5))
sns.countplot(x='Contract_Simple', hue='Churn', data=df)
plt.title('Churn by Contract Type (Simplified)')
plt.show()
```

```python
# Create a feature for internet service type
df['HasInternet'] = df['InternetService'].apply(lambda x: 'No' if x == 'No' else 'Yes')
```

```python
# Visualize churn by internet service presence
plt.figure(figsize=(8, 5))
sns.countplot(x='HasInternet', hue='Churn', data=df)
plt.title('Churn by Internet Service Presence')
plt.show()
```

# Step 5: Prepare Data for Modeling

```python
# Split into features and target
X = df_processed.drop('Churn', axis=1)
y = df_processed['Churn']
```

```python
# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Scale numerical features
scaler = StandardScaler()
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
```

```python
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])
```

```python
# Handle class imbalance with SMOTE
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
# Check new class distribution
print("Original class distribution:", np.bincount(y_train))
print("After SMOTE class distribution:", np.bincount(y_train_smote))
```

# Step 6: Model Building and Evaluation

```python
#Logistic Regression
```

```python
# Logistic Regression
logreg = LogisticRegression(max_iter=1000, random_state=42)
logreg.fit(X_train_smote, y_train_smote)
```

```python
# Predictions
y_pred = logreg.predict(X_test)
y_pred_proba = logreg.predict_proba(X_test)[:, 1]
```

```python
# Evaluation
print("Logistic Regression Results:")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nROC AUC Score:", roc_auc_score(y_test, y_pred_proba))
```

```python
# Feature importance
coef_df = pd.DataFrame({'Feature': X.columns, 'Coefficient': logreg.coef_[0]})
coef_df['Abs_Coefficient'] = coef_df['Coefficient'].abs()
coef_df = coef_df.sort_values('Abs_Coefficient', ascending=False)
print("\nTop 10 Important Features:")
print(coef_df.head(10))
```

```python
#Random Forest Classifier
```

```python
# Random Forest
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train_smote, y_train_smote)
```

```python
# Predictions
y_pred_rf = rf.predict(X_test)
y_pred_proba_rf = rf.predict_proba(X_test)[:, 1]
```

```python
# Evaluation
print("Random Forest Results:")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))
print("\nROC AUC Score:", roc_auc_score(y_test, y_pred_proba_rf))
```

```python
# Feature importance
feature_imp = pd.DataFrame({'Feature': X.columns, 'Importance': rf.feature_importances_})
feature_imp = feature_imp.sort_values('Importance', ascending=False)
print("\nTop 10 Important Features:")
print(feature_imp.head(10))
```

```python
# Plot feature importance
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_imp.head(20))
plt.title('Random Forest Feature Importance')
plt.show()
```

```python
# Hyperparameter tuning for Random Forest
```

```python
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE


rf_grid = RandomizedSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_distributions=param_grid,
    n_iter=10,             
    cv=3,                  
    scoring='roc_auc',     
    n_jobs=-1,             
    verbose=1,
    random_state=42
)
```

```python
rf_grid.fit(X_train_smote, y_train_smote)
```

```python
print("Best Parameters:", rf_grid.best_params_)
print("Best ROC AUC Score:", rf_grid.best_score_)
```

```python
# Evaluate best model
best_rf = rf_grid.best_estimator_
y_pred_best = best_rf.predict(X_test)
y_pred_proba_best = best_rf.predict_proba(X_test)[:, 1]
```

```python
print("\nTuned Random Forest Results:")
print("Accuracy:", accuracy_score(y_test, y_pred_best))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_best))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_best))
print("\nROC AUC Score:", roc_auc_score(y_test, y_pred_proba_best))
```

# Step 7: Model Interpretation and Business Insights

```python
# Analyze important features from the best model
feature_imp_best = pd.DataFrame({'Feature': X.columns, 'Importance': best_rf.feature_importances_})
feature_imp_best = feature_imp_best.sort_values('Importance', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_imp_best.head(20))
plt.title('Tuned Random Forest Feature Importance')
plt.show()
```

```python
# Analyze top features
top_features = feature_imp_best.head(10)['Feature'].values

for feature in top_features:
    if feature in ['tenure', 'MonthlyCharges', 'TotalCharges']:
        plt.figure(figsize=(10, 6))
        sns.boxplot(x='Churn', y=feature, data=df)
        plt.title(f'{feature} vs Churn')
        plt.show()
    else:
        # For categorical features
        original_feature = '_'.join(feature.split('_')[:-1])
        value = feature.split('_')[-1]
        
        if original_feature in df.columns:
            temp_df = df.copy()
            temp_df['Feature_Value'] = temp_df[original_feature].apply(lambda x: 1 if x == value else 0)
            
            plt.figure(figsize=(8, 5))
            sns.barplot(x='Feature_Value', y='Churn', data=temp_df)
            plt.title(f'Churn Rate for {original_feature} = {value}')
            plt.xticks([0, 1], [f'Not {value}', value])
            plt.show()
```
